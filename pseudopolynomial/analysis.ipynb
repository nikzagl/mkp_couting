```python
import pandas as pd
```

```python
for i in range(1,6):
    df = pd.read_csv(f"output_data/test_1/knap_d_{i}.txt", sep = ", ")
    print(f"d = {i}, mean time dynamic programming: {df['Time dynamic programming (ms)'].mean()} ms, mean time brute force: {df['Time brute force (ms)'].mean()} ms")
    print(f"d = {i}, std time dynamic programming: {df['Time dynamic programming (ms)'].std()} ms, std time brute force: {df['Time brute force (ms)'].std()} ms")
```

```python
for i in [5, 10, 15, 20]:
    df = pd.read_csv(f"output_data/test_2/n_{i}.txt", sep = ", ")
    print(f"d = {i}, mean time dynamic programming: {df['Time dynamic programming (mcs)'].mean()/1000} ms, std time dynamic programming(ms): {df['Time dynamic programming (mcs)'].std()/1000} ms")
    print(f"d = {i}, mean time brute force: {df['Time brute force (mcs)'].mean()/1000} ms, std time brute force: {df['Time brute force (mcs)'].std()/1000} ms")
```
