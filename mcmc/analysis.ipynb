```python
import pandas as pd

```

# Тест 1: 

```python
for i in [30, 45, 60, 75, 90]:
    df = pd.read_csv(f"output_data/test_1/b_0_{i}.txt")
    error_rate = (abs(df["Approximate result"]-df["Exact result"])/df["Approximate result"] >= 0.1).mean()
    time_approx_mean, time_approx_std = (df["Time approx"]/1000).mean(), (df["Time approx"]/1000).std() 
    time_exact = df["Time exact"][0]/1000
    print(f"b_0: {i}, time approx: {time_approx_mean} +- {time_approx_std} s, error rate: {error_rate}, time exact: {time_exact} s")
```
